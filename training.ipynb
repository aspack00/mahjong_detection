# Train model

```python
%load_ext tensorboard
%tensorboard --logdir runs/train
```

```python
import torch

model = YOLO("yolo11s.pt")
num_epochs = 20

# Use GPU if available, otherwise CPU; batch=-1 only makes sense with GPU
use_cuda = torch.cuda.is_available()
train_device = 0 if use_cuda else "cpu"
train_batch = -1 if use_cuda else 8  # smaller batch on CPU to avoid OOM

results = model.train(
    data=data_root_path.absolute() / "data.yaml",
    epochs=num_epochs,
    batch=train_batch,
    device=train_device,
    imgsz=1024,
)
```
